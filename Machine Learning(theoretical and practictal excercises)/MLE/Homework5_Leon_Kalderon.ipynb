```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import math
%matplotlib inline

#(A)
A = mpimg.imread('huble_tel.jpg')
plt.imshow(A)
plt.show()
```

```python
#(b)-(c)

#store the dimenensions of the image
M = np.shape(A)[0]
N = np.shape(A)[1]
#store how many elements we want to generate
N_elems = M * N
#mean and variance of the normal distribution that we will generate our values
mu, sigma = 0, 1 
#B is a list of 15 matrices with gaussian noise.
B = []
#C is a list of 15 matrices with A + gaussian noise
C = []
#B_avg is a matrix with the average noise between all the Bs matrices
B_avg = np.zeros((M, N))

for i in range(0,15):
    #parameteres mean, standard deviation, number of points
    elements = np.random.normal(mu, sigma, N_elems)
    #build a temporary matrix of MxN dimenensions with the element that we generated above
    temp = np.reshape(elements, (M, N))
    #append the matrix to the list multiplied by 259^(1/2)
    B.append(temp * pow(259,(1/2)) )
    #build the matrix with the average noise
    B_avg = B_avg + temp * (1/15) 
    C.append(A+B[i])
    
```

```python
#Here we plot all the 15 versions of A + noise
for i in range(0,15):
    plt.imshow(C[i])
    plt.show()
#We observe that the pixel's color are slightly changed because of the noise we added on the matrices
```

```python
#Now we will plot the C_avg
C_avg = (B_avg) * pow(259,(1/2)) + A
plt.imshow(C_avg)
plt.show() 
```

We observe that the image that have been produced is (almost) the same with the A. 
This make sense because when we average the noises that have been generated from a normal distribution with mean=0, they converge to 0. (E[] ~ 0).
