```python
#A
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt


Data = sio.loadmat('regression.mat')
X = Data['X']
y = Data['y']
X
```

```python
#B
plt.plot(X, y, 'o', label='Original data', markersize=10)
plt.show()
```

```python
#E
X_square = np.power(X[0],2)

X_cube = np.power(X[0],3)

X_stack = np.stack((np.ones((len(X[0]))), X[0], X_square, X_cube),axis=-1)

X_inverse = np.linalg.inv(np.dot(X_stack.T,X_stack))

theta = X_inverse.dot(X_stack.T).dot(y[0])
theta
X_inverse.dot(X_stack.T)
```
