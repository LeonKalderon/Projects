```python
#h)
#we create a new DF with the data that we want to use
df = sumdf.iloc[:,0:27].set_index('chimpcode').loc[:,'dom':'innov'].copy()

#Standardized version:
#I also tried to use the code below to standardize the dataframe before comparing the rows; but the results were the same.
#That happens because the 24 personality traits were rated on the same scale.
''''df_normalized = df.apply(lambda col: col.apply(lambda row: (row - col.mean())/col.std()) , axis = 0)'''

#We create python lists to store the distances and similarities between chimpanzees.
#We use lists ,instead of panda's dataframe, because it is much faster 
#to add values into lists than adding a new row to a new DF on each itteration.(~100x faster after we checked the runtimes with %%timeit)
#In the end, we can append these lists to a DF, so we can manipulate better the results  
pairs,eucl,manh,cos=([] for i in range(4))

#These functions are 3 different ways to find the distances between 2 vectors.
#We will use them all to compare the results
def manhattan_dist(x,y):
    return sum(abs(a-b) for a,b in zip(x,y))

def euclidean_dist(x,y):
    return (sum(pow(a-b,2) for a, b in zip(x, y)))**(1/2)

def cosine_sim(x,y):
    numerator = sum(a*b for a,b in zip(x,y))
    denominator = square_rooted(x)*square_rooted(y)
    return round(numerator/float(denominator),3)

def square_rooted(x):
    return round(math.sqrt(sum([a*a for a in x])),3)

#We compare all rows with each other
for code1, val1 in df.iterrows():
    #We slice the DF from the current row to skip duplicate comparisons.  
    for code2, val2 in df.loc[code1:].iterrows():
        if code1 != code2:
            pairs.append(code1 +'-'+ code2)
            eucl.append(euclidean_dist(val1, val2))
            manh.append(manhattan_dist(val1, val2))
            cos.append(cosine_sim(val1, val2))

#We create a dataframe with all the results         
distdf = pd.DataFrame(
    {'pair': pairs,
     'euclidean_dist': eucl,
     'manhattan_dist': manh,
     'cosine_sim':cos
    }).set_index('pair')

#We look for the max cosine_similairy value, because this value shows how similar the 2 vectors are
#in correspondance, we search for the minimum value for the other two methods, because we want to find the minimum distance. 
distdf['cosine_sim'].idxmax(axis = 0)         #result: 'Q217-D224'
distdf['euclidean_dist'].idxmin(axis = 0)     #result: 'Q217-D224'
distdf['manhattan_dist'].idxmin(axis = 0)     #result: 'Q217-D224'   
```
