```python
from bs4 import BeautifulSoup
import requests

#In the variable author we assign the name of the author we want to search for his/her books.(e.g. Oscar Wilde)
#This script works for any author. Be careful!Author's name is 'space' sensitive.(e.g. ' CharlesDickens  ' won't work) 
#Some books exist more than once because different versions of these books are uploaded on the site.
author = 'Charles Dickens'
books = []
author_to_URL = author.replace(' ', '+')
query_url = '/ebooks/search/?query=' + author_to_URL
URL = 'http://www.gutenberg.org' + query_url

    
def collect_books(link):
    #Error handling
    try:
        r = requests.get(link)
    except requests.exceptions.MissingSchema as ms: # Missing URL schema
        print(ms)
    except requests.exceptions.ConnectionError as ce: # connection error
        print(ce)
    except requests.exceptions.HTTPError as herror: # invalid HTTP response
        print(herror)
    except requests.exceptions.Timeout as toerr:
        print(toerr)
    else: #Page retrieval OK.
        html = r.content
        #We use 'html5lib' parser instead of 'html.parser' that sometimes skips wanted tags. This parser is more consistent.
        soup = BeautifulSoup(html, 'html5lib')
        #We collect tags that contain current webpage's books and audiobooks
        booklinks = soup.find_all('li', {'class': 'booklink'})
        for booklink in booklinks:
            #we check if the tag contains a audiobook so we can exclude it from the list.
            if booklink.find('span', {'class': 'icon icon_audiobook'}) is not None: 
                continue
            else:
                title = booklink.find('span', {'class': 'title'}).get_text()
                #we check if the book has an assigned author and then we check if the author is the one we queried for
                if booklink.find('span', {'class': 'subtitle'}) is not None:
                    subtitle = booklink.find('span', {'class': 'subtitle'}).get_text()
                    if author in subtitle:
                        books.append(title)
                        print(title)
        
        #We find the next page's URL and if it exists we call recursivly the function    
        next_link = soup.find('a', {'title': 'Go to the next page of results.'})
        if next_link is not None:
            link = 'http://www.gutenberg.org' + next_link.get('href')
            collect_books(link)
      


    
collect_books(URL)
```
