```python
import requests
import re

#Retrieve the text from the URL and store it into the variable book.
URL = 'http://www.gutenberg.org/cache/epub/30368/pg30368.txt'
book = requests.get(URL).text
#Define a function that prints the sentences of a text that contain a certain word.
def print_sent_contain_word(text, word):
    count = 0
    #Remove text in '[..]'
    text = re.sub(r'\[.*?\]', '', text)
    #Split the text into sentences and store them into the list 'sentences'.
    #Sentences end with ." , ?" , !" will not be splitted with the regex below.
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)(\s|[A-Z].*)', text)
    #We iterate through the list of sentences and we check if they contain the word that we are looking for.
    for s in sentences:
        #Strip spaces/tabs/newlines
        s = ' '.join(s.split())
        #We split the setences that was not splitted properly.
        sent2 = re.split(r'[.?!]\"', s)
        #We print the sentences
        for se in sent2:
            if word in se:
                count +=1
                print('Sentence ',count, ':')
                print(se)

print_sent_contain_word(book, 'Scrooge')
```
